# Clase Publica

```python
class Personaje:
    # Metodo constructor
    def __init__(self, nombre, clase, nivel, saluda, mana):
        self.nombre = nombre
        self.clase = clase
        self.nivel = nivel
        self.salud = saluda
        self.mana = mana

    # Metodo atacar
    def atacar(self, objetivo):
        return f"{self.nombre} ataca a {objetivo}."

    # Metodo curar
    def curar(self, cantidad):
        self.salud += cantidad
        return f"{self.nombre} se cura a sí mismo por {cantidad} puntos de salud."
```

```python
# Instanciación de objetos

# Instanciación de objetos se refiere al proceso de crear un objeto a partir de una clase. Cuando instancias un objeto, estas creando una instancia de esa clase con atributos y métodos definidos
heroe = Personaje("Aragorn", "Guerrero", 20, 100, 50)

heroe.atacar("Horco")
```

```python
heroe.curar(20)
```

```python
heroe.nombre
```

# Clase Privada / Encapsulacion

```python
class Personaje_2:
    # Metodo constructor
    def __init__(self, nombre, clase, nivel, saluda, mana):
        self.__nombre = nombre
        self.__clase = clase
        self.__nivel = nivel
        self.__salud = saluda
        self.__mana = mana

    # Metodo atacar
    def atacar(self, objetivo):
        return f"{self.__nombre} ataca a {objetivo}."

    # Metodo curar
    def curar(self, cantidad):
        self.__salud += cantidad
        return f"{self.__nombre} se cura a sí mismo por {cantidad} puntos de salud."

    def obtener_nombre(self): # metodo getter para obtener el nombre del personaje
        return self.__nombre

    def obtener_nivel(self): # metodo getter para obtener el nivel del personaje
        return self.__nivel

    def set_nivel(self, nuevo_nivel): # metodo setter para modificar el nivel del personaje
        self.__nivel = nuevo_nivel
```

```python
heroe_2 = Personaje_2("Aragorn", "Guerrero", 20, 100, 50)

heroe_2.obtener_nombre()
```

```python
heroe_2.set_nivel(25)
heroe_2.obtener_nivel()
```

# Herencia

```python
class Personaje_3:
    # Metodo constructor
    def __init__(self, nombre, clase, nivel, saluda, mana):
        self.nombre = nombre
        self.clase = clase
        self.nivel = nivel
        self.salud = saluda
        self.mana = mana

    # Metodo atacar
    def atacar(self, objetivo):
        return f"{self.nombre} ataca a {objetivo}."

    # Metodo curar
    def curar(self, cantidad):
        self.salud += cantidad
        return f"{self.nombre} se cura a sí mismo por {cantidad} puntos de salud."
```

```python
# Clase derivada
class Mago(Personaje_3):
    def __init__(self, nombre, nivel, salud, mana, poder_magico):
        super().__init__(nombre, "Mago", nivel, salud, mana)
        self.poder_magico = poder_magico

    def lanzar_hechizo(self, hechizo, objetivo):
        if self.mana > 0:
            self.mana -= 10
            return f"{self.nombre} lanza {hechizo} a {objetivo} con poder {self.poder_magico}!"
        else:
            return f"{self.nombre} no tiene suficiente mana para lanzar {hechizo}."
```

```python
mago = Mago("Gandalf", 30, 80, 100, 50)
print(mago.atacar("Orco"))
```

```python
print(mago.lanzar_hechizo("Bola de Fuego", "Orco"))
```

# Sobreescritura de metodos

```python
class Guerrero(Personaje_3):
    def __init__(self, nombre, nivel, salud, mana, fuerza):
        super().__init__(nombre, "Guerrero", nivel, salud, mana)
        self.fuerza = fuerza

    def atacar(self, objetivo):
        return f"{self.nombre} ataca a {objetivo} con fuerza {self.fuerza}!"
```

```python
mago = Mago("Gandalf", 50, 120, 200, 80)
guerrero = Guerrero("Aragorn", 40, 150, 50, 70)
print(mago.atacar("Orco"))
```

```python
print(guerrero.atacar("Orco"))
```

# Polimorfismo

```python
# Uso del polimorfismo
def realizar_ataque(personaje, objetivo):
    print("Te lanzaré mi mejor golpe")
    print(personaje.atacar(objetivo))
```

```python
realizar_ataque(mago, "Orco")
```

```python
realizar_ataque(guerrero, "Orco")
```

# Herencia multiple

```python
class Personaje_4:
    def __init__(self, nombre, nivel, salud):
        self.nombre = nombre
        self.nivel = nivel
        self.salud = salud

    def atacar(self, objetivo):
        return f"{self.nombre} ataca a {objetivo}!"

# Clase base 2
class Hechicero_2:
    def __init__(self, mana, poder_magico):
        self.mana = mana
        self.poder_magico = poder_magico

    def lanzar_hechizo(self, hechizo, objetivo):
        if self.mana > 0:
            self.mana -= 10
            return f"El hechicero lanza {hechizo} a {objetivo} con poder {self.poder_magico}!"
        else:
            return "El hechicero no tiene suficiente mana para lanzar hechizos."
```

```python
# Clase derivada con herencia múltiple
class GuerreroMago(Personaje_4, Hechicero_2):
    def __init__(self, nombre, nivel, salud, mana, poder_magico, fuerza):
        Personaje_4.__init__(self, nombre, nivel, salud)
        Hechicero_2.__init__(self, mana, poder_magico)
        self.fuerza = fuerza
    
    def atacar_con_fuerza(self, objetivo):
        return f"{self.nombre} ataca a {objetivo} con fuerza {self.fuerza}!"
```

```python
guerreroMago = GuerreroMago("Elrond", 45, 110, 150, 60, 80)
print(guerreroMago.atacar("Orco"))
```

```python
print(guerreroMago.lanzar_hechizo("Bola de Fuego", "Orco"))
```

```python
print(guerreroMago.atacar_con_fuerza("Orco"))
```

# Metodos mágicos

```python
class Personaje_5:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad
```

```python
persona1 = Personaje_5("Frodo", 50)
persona1.nombre
```

```python
print(persona1)
```

```python
class Persona_6:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def __str__(self):
        return f"Persona: {self.nombre}, Edad: {self.edad}"

    def __repr__(self):
        return f"Persona(nombre='{self.nombre}', edad={self.edad})"

    def __add__(self, otra_persona):
        return f"{self.nombre} y {otra_persona.nombre} son amigos."
    
    def __len__(self):
        return self.edad
```

```python
persona2 = Persona_6("Frodo", 50)
print(persona2)
```

```python
print(repr(persona2))
```

```python
persona2 = Persona_6("Argan", 70)
```

```python
print(persona2 + persona1)
```

```python
print(len(persona2))
```
